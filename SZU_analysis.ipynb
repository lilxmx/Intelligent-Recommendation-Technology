```python
import matplotlib.pyplot as plt

# 设置字体以避免中文显示为乱码
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负数的负号显示问题

# 定义数据
labels = ['290-300', '300-310', '310-320', '320-330', '330-340', '340-350', '350-360', '360-370', '370-380']
initial_counts_values = [3, 11, 19, 19, 14, 11, 11, 6, 2]
admission_counts_values = [2, 2, 9, 14, 12, 10, 10, 6, 2]
admission_ratio_values = [0.666667, 0.181818, 0.473684, 0.736842, 0.857143, 0.909091, 0.909091, 1.000000, 1.000000]

# 设置柱状图位置
x = range(len(labels))

fig, ax1 = plt.subplots(figsize=(12, 8))

# 绘制初试成绩人数和录取人数的柱状图
bar_width = 0.35
ax1.bar(x, initial_counts_values, width=bar_width, label='初试成绩人数', align='center', alpha=0.7)
ax1.bar([i + bar_width for i in x], admission_counts_values, width=bar_width, label='录取人数', align='center', alpha=0.7)

# 设置轴标签和标题
ax1.set_xlabel('分数段')
ax1.set_ylabel('人数')
ax1.set_title('各分数段的初试成绩人数和录取人数')
ax1.set_xticks([i + bar_width / 2 for i in x])
ax1.set_xticklabels(labels, rotation=45)

# 创建第二个y轴
ax2 = ax1.twinx()
ax2.plot([i + bar_width / 2 for i in x], admission_ratio_values, color='red', marker='o', label='录取比例')

# 设置第二个y轴的标签
ax2.set_ylabel('录取比例')
ax2.set_ylim(0, 1.1)

# 添加图例
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# 显示图形
plt.tight_layout()
plt.show()
```
